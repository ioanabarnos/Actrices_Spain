## Actores españoles
Extraiga de IMDB los actores nacidos en España.

Pasos:

* 1 - Localizamos la web de IMDB: https://www.imdb.com.
* 2 - Vamos a la seccion de búsqueda: https://www.imdb.com/search/name/?ref_=fn_asr_nm.
* 3 - Jugamos con los parámetros para localizar la URL: https://www.imdb.com/search/name/?birth_place=spain&gender=female&count=100.

En los parámetros vemos que el nacimiento lo hemos puesto en "spain", hemos activado unicamente la casilla "female" en gender identity y el número de registro está situado en 100.

* 4 - Imprima la lista de nombres en una tabla junto con la pequeña descripción (Bio) y su foto
* 5 - Recuerde que los datos se encuentra en la parte HTML "div" y la clase correspondiente es "lister-item mode-detail"


```python
# importamos las librerias
import requests
import bs4
import pandas as pd
import csv
```

```python
# creamos la variable "url" con el enlace al que queremos que acceda
URL = 'https://www.imdb.com/search/name/?birth_place=spain&gender=female&count=100'
```

```python
# comprobamos que puede acceder a la página con requests.get
response = requests.get(URL)
print(response)
```

```python
# creamos una variable "texto" a partir del texto en bruto extraido de la URL
texto = requests.get(URL).text
print(texto)
```

```python
# creamos la variable "soup" con BeautifulSoup para depurar
soup = bs4.BeautifulSoup(texto, 'html.parser')
print(soup)
```

```python
# La información que queremos se encuentra en la parte HTML "div"
    # extraemos las tablas a partir del código indicando cual es su clase
actrices = soup.find_all("div" , class_ = "lister-item mode-detail")
```

```python
# Cada uno de los actores los tenemos en las distintas posiciones de la variable "tablas"
actrices[0]
```

```python
# Nos quedamos con el "Nombre", "Foto" y "Bio"
nombres = []
imagenes = []
bios = []

# Iteramos a través de las posiciones del texto para sacar cada uno de los elementos

for actriz in actrices:
    
    # Nombre
    nombre = actriz.h3.a.text.strip()
    nombres.append(nombre)
    
    # Imagen
    imagen = actriz.find('img')['src']
    imagenes.append(imagen)
    
    # Bio
    bio = actriz.p.a.text.strip()
    bio = actriz.text.split(bio)[1].strip()
    bio = list(bio.split("."))[0]
    bios.append(bio)
```

```python
actrices_df = pd.DataFrame({'Nombre': nombres,'Imagen': imagenes,'Bio': bios})
print(actrices_df.info())
actrices_df.head(10)
```

```python
# Para imprimir la foto, usamos el siguiente código:
from IPython.core.display import HTML

# Convertimos una URL en una imagen
def path_to_image_html(path):
    return '<img src="'+ path + '" width="60" >'
```

```python
# imprimimos el dataframe con las imagenes definidas
HTML(actrices_df.to_html(escape=False ,formatters=dict(Imagen=path_to_image_html)))
```
